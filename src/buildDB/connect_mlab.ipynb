```python
from pymongo import MongoClient
import os

url = os.environ['mlab_db_link']

client = MongoClient(url)

db = client.get_database()
```

```python
db.history.count()
```

```python
price = db['price']
```

```python
import time
from datetime import datetime
time_format = "%Y%m%d"
```

```python
stocks = price.find({})
```

```python
for stock in stocks:
    if stock['history'][-1]['date'] != datetime.strptime(time.strftime(time_format), time_format):
        print(stock['code'])
```
